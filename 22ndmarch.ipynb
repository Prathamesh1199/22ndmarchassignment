```python
1:
  To calculate the Pearson correlation coefficient,we  will need to first compute the covariance 
and standard deviation of the two variables. Once you have these values, you can use the following
formula to calculate the Pearson correlation coefficient:

r = cov(X,Y) / (SD(X) * SD(Y))

Where:

.r is the Pearson correlation coefficient
.cov(X,Y) is the covariance between the two variables X and Y
.SD(X) is the standard deviation of X
.SD(Y) is the standard deviation of Y

Once you have calculated the Pearson correlation coefficient, you can interpret the result as follows:

.If r is positive, it indicates a positive linear relationship between the two variables. 
This means that as one variable increases, the other variable also tends to increase. 
The closer r is to +1, the stronger the positive correlation.
.If r is negative, it indicates a negative linear relationship between the two variables. 
This means that as one variable increases, the other variable tends to decrease. The closer
r is to -1, the stronger the negative correlation.
.If r is close to 0, it indicates that there is little or no linear relationship between the two
variables. This means that changes in one variable are not associated with changes in the other variable.


       In the context of the example provided, a positive Pearson correlation coefficient would indicate that there
is a positive relationship between the amount of time students spend studying for an exam and their final 
exam scores. In other words, as students spend more time studying, their exam scores tend to increase.
Conversely, a negative Pearson correlation coefficient would indicate a negative relationship between these 
variables, meaning that as students spend less time studying, their exam scores tend to decrease. If the Pearson 
correlation coefficient is close to 0, it would indicate that there is little or no relationship between the two variables.


```

```python
2:
  To calculate Spearmans rank correlation coefficient,we will need to first rank the values
for each variable, from smallest to largest. Once you have the ranked values, you can calculate 
the difference between the ranks for each observation, and then compute the Spearman's rank 
correlation coefficient using the following formula:

ρ = 1 - (6∑d^2) / (n(n^2 - 1))

Where:

.ρ is the Spearman's rank correlation coefficient
.d is the difference between the ranks for each observation
.n is the number of observations

Once you have calculated the Spearmans rank correlation coefficient, you can interpret the result as follows:

.If ρ is positive, it indicates a monotonic increasing relationship between the two variables. This means that as
one variable increases, the other variable tends to increase as well, but not necessarily at a constant rate. The 
closer ρ is to +1, the stronger the positive rank correlation.
.If ρ is negative, it indicates a monotonic decreasing relationship between the two variables. This means that as one
variable increases, the other variable tends to decrease, but not necessarily at a constant rate. The closer ρ is to -1, 
the stronger the negative rank correlation.
.If ρ is close to 0, it indicates that there is little or no monotonic relationship between the two variables.

  In the context of the example provided, a positive Spearmans rank correlation coefficient would indicate a monotonic 
increasing relationship between the amount of sleep individuals get each night and their overall job satisfaction level. 
In other words, as individuals get more sleep, their job satisfaction tends to increase, but not necessarily at a constant rate.
Conversely, a negative Spearmans rank correlation coefficient would indicate a monotonic decreasing relationship between these
variables, meaning that as individuals get less sleep, their job satisfaction tends to decrease, but not necessarily at a constant rate. 
If the Spearmans rank correlation coefficient is close to 0, it would indicate that there is little or no monotonic relationship between the two variables.

```
